## Recap: Efficient Taylor Series (from Lecture 1)

We start by revisiting the efficient computation of `exp(x)` using the recurrence relation for the Taylor series.

This will serve as a bridge to defining reusable functions.

````
# Efficient Taylor series for exp(x) – from Lecture 1
import math

s2 = 1
x = 0.123
t = x          # first term t_1 = x
n = 25
for j in range(1, n+1):
    if j % 15 == 0 or j == 1:
        print(f'j={j}, term t_j={t:.6e}')
    s2 += t
    t = x * t / (j+1)   # recurrence: t_{j+1} = x * t_j / (j+1)

print(f'Computed exp({x}) = {s2:.6f}, true = {math.exp(x):.6f}')
````

## Task 8: From Functions to Classes

### 1. Simple Function (`def`)
We start by encapsulating the Taylor series computation into a reusable function.

````
print('Task 8 – from "def" to advanced "class"')
print("First, a simple function (similar to MATLAB's function):")

def my_sum(x, n):
    """Compute sum_{k=0}^{n} x^k/k! using efficient recurrence."""
    s = 1.0
    t = x
    for j in range(1, n+1):
        s += t
        t = x * t / (j+1)
    return s

x = 0.123
answer = my_sum(x, 25)
print(f'my_sum({x}, 25) = {answer:.6f}, true exp = {math.exp(x):.6f}')
````

### 2. Introduction to Classes

A **class** groups data (attributes) and functions (methods) together. This is the foundation of **Object-Oriented Programming**.

We will now implement the same functionality using a class, with three different styles to illustrate flexibility.

#### Class Form 1: Store `x` and `n` as attributes

````
print('Class form 1: Attributes set in __init__')

class my_sum1:
    def __init__(self, x=1.0, n=10):
        self.x = x
        self.n = n
    
    def comp(self):
        s = 1.0
        t = self.x
        for j in range(1, self.n + 1):
            s += t
            t = self.x * t / (j + 1)
        return s

# Usage
obj = my_sum1(x=0.789, n=15)
ans1 = obj.comp()
print(f'my_sum1(0.789, 15) = {ans1:.6f}, true exp = {math.exp(0.789):.6f}')

x = 0.123
ans = my_sum1(x, 25).comp()
print(f'my_sum1({x}, 25) = {ans:.6f}, true exp = {math.exp(x):.6f}')
````

#### Class Form 2: `x` is stored, `n` is passed to the method

````
print('Class form 2: x stored, n passed to comp()')

class my_sum2:
    def __init__(self, x=1.0):
        self.x = x
    
    def comp(self, n=10):
        s = 1.0
        t = self.x
        for j in range(1, n + 1):
            s += t
            t = self.x * t / (j + 1)
        return s

# Usage
obj2 = my_sum2(x=0.789)
ans2 = obj2.comp(n=10)
print(f'my_sum2(0.789, 10) = {ans2:.6f}')

obj3 = my_sum2(x=0.123)
ans3 = obj3.comp(25)
print(f'my_sum2(0.123, 25) = {ans3:.6f}, true exp = {math.exp(0.123):.6f}')
````

#### Class Form 3: No attributes – everything passed to the method

````
print('Class form 3: no attributes, parameters passed to comp()')

class my_sum3:
    def __init__(self):
        pass
    
    def comp(self, x=1.2, n=10):
        s = 1.0
        t = x
        for j in range(1, n + 1):
            s += t
            t = x * t / (j + 1)
        return s

# Usage: note that named arguments can be swapped
ans4 = my_sum3().comp(n=10, x=0.789)
print(f'my_sum3(0.789, 10) = {ans4:.6f}')

x = 0.123
ans5 = my_sum3().comp(x, 25)
print(f'my_sum3({x}, 25) = {ans5:.6f}, true exp = {math.exp(x):.6f}')
````

**Key Takeaways:**
- `__init__` is the constructor – it initialises the object's attributes.
- `self` refers to the current instance; it must be the first parameter of any method.
- You can design classes in multiple ways – choose based on how you want to use them.
- Classes are powerful for grouping data and functionality, especially in larger projects.

## Task 9: Reading All Images from a Directory Using `glob`

In many deep learning tasks, you need to load a batch of images. The `glob` module helps you find all files matching a pattern.

We will demonstrate this with a synthetic dataset (assuming the directories exist). If they don't, the code will gracefully inform you.

````
import warnings
warnings.simplefilter("ignore", UserWarning)

import glob
import os
import matplotlib.pyplot as plt
from skimage import io

print("Task 9 – Reading images from a directory using glob")

# Define wildcard patterns (adjust paths to your own dataset)
str_img1 = '../Cells/train/image/*.png'
str_lab1 = '../Cells/train/label/*.png'

# Check if the image directory exists
img_dir = os.path.dirname(str_img1)
if not os.path.exists(img_dir):
    print(f'Directory "{img_dir}" does not exist. Skipping image loading.')
    print('To test this, create the folder and place some PNG images there.')
else:
    # Get sorted list of image files
    files_img = sorted(glob.glob(str_img1, recursive=True))
    files_lab = sorted(glob.glob(str_lab1, recursive=True))
    
    print(f'Found {len(files_img)} image files and {len(files_lab)} label files.')
    
    if len(files_img) != len(files_lab):
        print('Warning: number of images and labels do not match!')
    
    # Display a few images
    for i in range(min(len(files_img), 10)):  # show at most 10
        img = io.imread(files_img[i], 0)  # grayscale
        if i % 7 == 0:   # display every 7th
            plt.imshow(img, cmap='gray')
            plt.title(f'Image {i}')
            plt.draw()
            plt.pause(0.5)
    
    plt.figure(figsize=(8, 4))
    # Show image and its label side by side for the first few
    for i in range(min(3, len(files_img))):
        img = io.imread(files_img[i], 1)  # colour
        lab = io.imread(files_lab[i], 0)  # label (grayscale)
        plt.subplot(1, 2, 1)
        plt.imshow(img)
        plt.title(f'Image {i}')
        plt.subplot(1, 2, 2)
        plt.imshow(lab, cmap='gray')
        plt.title(f'Label {i}')
        plt.show()
        
    print('Finished displaying images.')
````

**Note:** The above code assumes a folder structure common in medical imaging datasets. Adjust the paths to your own data.

The key functions:
- `glob.glob(pattern)` returns a list of all filenames matching the pattern.
- `io.imread()` from `skimage` reads images into NumPy arrays.
- Using `sorted()` ensures a deterministic order.
