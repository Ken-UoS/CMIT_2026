## Part 3: Math Functions & User-Defined Functions

Python's `math` module provides many useful mathematical tools. You can also define your own functions using `def`.

```python
print('\nTask 3 --- calling a built-in math function or defining own')
import math

a = math.exp(1.0)   # e^1
b = math.pi
print('math.exp(1)=', a, ', math.pi=', b, ', math.sinh(1)=', math.sinh(1.0))

# Custom implementation of hyperbolic sine
def my_sinh(x):
    y = math.exp(x)
    ym = math.exp(-x)
    return (y - ym) / 2.0

print('my_sinh(1.0)=', my_sinh(1.0), '(compare with the built-in function)')
```

## Part 4: Conditional Statements (`if`)

Conditionals allow your program to make decisions. Remember: **indentation** is the core syntax for defining code blocks in Python.

```python
print('\nTask 4 --- Programming basics: (i) "if" conditional')
import numpy as np
from numpy.random import rand as rd

x = rd() - 0.5   # Random number in [-0.5, 0.5)
y = np.random.rand() - 0.5

print('x=', x, ' y=', y)

# Standard if-else
if x >= 0.0:
    print('non-negative x')
else:
    print('\tnegative x')

if y >= 0.1:
    print('positive y and y>=0.1')
else:
    print('\tless than 0.1')
print('|')

# Python's special ternary operator (inline if-else)
x1 = 'non-negative x' if x >= 0.0 else '\tnegative x'
y1 = 'positive y and y>=0.1' if y >= 0.1 else '\tless than 0.1'
print(x1, ' | ', y1)
```

## Part 6: `for` Loops – Taylor Series for `exp(x)`

Loops are fundamental to programming. Here, we compute the Taylor series of `e^x` using two different approaches to compare efficiency and accuracy.

```python
print('\nTask 6 --- Programming basics: (ii) "for" loop')

print('\tfor -- example 1: Direct call to math.factorial')
n = 8
s1 = 1.0
x = 0.123
for i in range(1, n+1):
    s1 += x**i / math.factorial(i)   # Alternatively: s1 = s1 + ...
print('s1 =', s1, ' (n=', n, ')  True exp(x)=', math.exp(x))
```

The method above recomputes `x**i` and `factorial(i)` each time. A more efficient way is to use **recurrence**.

```python
print('\n\tfor -- example 2: Efficient recurrence (using the previous term)')
s2 = 1.0
x = 0.123
t = x          # First term t_1 = x
n = 25
for j in range(1, n+1):
    if j % 5 == 0 or j == 1:   # Print every 5 steps to reduce clutter
        print('j=', j, ' term t_j=', t)
    s2 += t
    t = x * t / (j+1)          # Recurrence: t_{j+1} = x * t_j / (j+1)
print('When x=', x, ',  s2 = %.8f' % s2, 'NB exp(x) = %.8f' % math.exp(x),
      ' error = {:.2e}'.format(s2 - math.exp(x)))
```

**Recurrence Principle**:
For the Taylor expansion `e^x = Σ x^k/k!`, let the k-th term be `t_k = x^k/k!`.
Then `t_{k+1} = t_k * x / (k+1)`. This avoids expensive power and factorial computations inside the loop.

## Part 7: NumPy Arrays & PyTorch Tensors

Scientific computing relies heavily on arrays. **NumPy** provides the foundation, while **PyTorch** extends this for deep learning.

```python
print('\nTask 7 --- Programming basics: (iii) arrays, matrices, tensors')
import numpy as np

print('\n\t vectors a b c')
a = np.arange(4)          # [0,1,2,3]
b = np.arange(2, 7)       # [2,3,4,5,6]
print('a=', a, 'b=', b)

h = 0.125
c = np.arange(3.0, 4.0+h, h)   # Start, stop (inclusive), step
print('c =', c)
print('Length: a b c =', len(a), len(b), len(c), ' and shape a =', a.shape)

# Random matrix
ar = np.random.rand(3, 5)
print('ar =\n', ar)
print('HERE ar shape =', ar.shape, ' after conversion a`s shape =',
      a[:, np.newaxis].shape, 'or to row =', a[np.newaxis, :].shape)

# Slicing arrays
a1 = ar[0:2, 0:2]
a2 = ar[-2:, -2:]
print('ar`s submatrices =\n', a1, ' and\n', a2)

# 3D Tensor
at = np.random.rand(3, 5, 2)
print('HERE at (tensor) shape =', at.shape, ' after conversion at`s shape =',
      at[np.newaxis, :].shape)
```

## 🏠 Homework 1: The Hilbert Matrix and Its Condition Number

This is your first programming assignment for this course. Try to complete it on your own.

```python
# TASK: Write a code to generate the Hilbert matrix H with H_{ij}=1/(i*j - 1)
# for i,j = 1,...,n. Then compute its condition number cond(H) for n=4 and n=8
# using numpy.linalg.cond.
# Show your answers (print them) in the next lecture.

# --- Your code here ---
import numpy as np

def hilbert(n):
    # Implement the Hilbert matrix
    # Hint: use nested loops or vectorized operations
    H = np.zeros((n, n))
    for i in range(1, n+1):
        for j in range(1, n+1):
            H[i-1, j-1] = 1.0 / (i*j - 1)
    return H

# Compute for n=4 and n=8
for n in [4, 8]:
    H = hilbert(n)
    cond = np.linalg.cond(H)
    print(f"Condition number for n={n}: {cond:.3e}")

# Bonus (vectorized version):
# i = np.arange(1, n+1).reshape(-1,1)
# j = np.arange(1, n+1).reshape(1,-1)
# H = 1.0 / (i*j - 1)
# print(np.linalg.cond(H))
```
